```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from google.colab import files
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
uploaded=files.upload()
file_path='1651277648862_healthinsurance.csv'
df=pd.read_csv(file_path)
```

```python
# Drop ID columns if present
# df.drop(columns=['ID'], inplace=True)  # Uncomment if there's an ID column

# Convert categorical columns to dummy variables
df_encoded = pd.get_dummies(df, drop_first=True)

# Normalize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded)
```

```python
df_filled = df.fillna(df.mean(numeric_only=True))  # Fill numeric NaNs with mean
```

```python
for col in df_filled.select_dtypes(include=['object']).columns:
    df_filled[col] = df_filled[col].fillna(df_filled[col].mode()[0])
```

```python
X = df_filled.drop(columns=['claim'])
X = pd.get_dummies(X, drop_first=True)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_data)
    inertia.append(km.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid()
plt.show()

```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)  # Use scaled features

# Get cluster labels
cluster_labels = kmeans.labels_
```

```python
# Add the cluster labels to your original DataFrame
df['Cluster'] = cluster_labels
```

```python
# Check how many samples in each cluster
print(df['Cluster'].value_counts())

# Get average values per cluster
print(df.groupby('Cluster').mean(numeric_only=True))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce to 2 components for plotting
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

# Plot clusters
plt.figure(figsize=(8,6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=cluster_labels, cmap='viridis')
plt.title('KMeans Clustering (k=3)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```
